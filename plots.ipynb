```python
import pandas as pd
import seaborn as sns
data = pd.read_excel("sheets/stocks_data.xlsx", index_col=0)
daily_return = data.pct_change().round(4)
```

```python
data.plot(subplots=True, figsize=(10, 10), title="Stock Prices", grid=True)
```

```python
daily_return.plot(subplots=True, figsize=(10, 10), title="Daily Returns", grid=True)
```

```python
(1+daily_return).cumprod().round(7).plot()
```
